```python
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
for index, word in enumerate(wv.index2word):
    if index == 10:
        break
    print(f"word #{index}/{len(wv.index2word)} is {word}")
```

```python
pairs = [
    ('car', 'Ferrari'),   # a minivan is a kind of car
    ('car', 'Wheels'),   # still a wheeled vehicle
    ('car', 'Television'),  # ok, no wheels, but still a vehicle
    ('car', 'Iran'),    # ... and so on
    ('car', 'India'),
]
for w1, w2 in pairs:
    print('%r\t%r\t%.2f' % (w1, w2, wv.similarity(w1, w2)))
```
